```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# o numpy da suporte ao pandas, mas não é necessario usar muita coisa dele
# o comando abaixo gera uma lista
x = np.linspace(1,10,100) 
```

```python
x
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
np.sin(1)
```

```python
# o 3º parametro é a cor da linha 
plt.plot(x, np.sin(x), 'r')
plt.show()
```

```python
plt.plot(x, np.sin(x), 'r--')
plt.show()
```
