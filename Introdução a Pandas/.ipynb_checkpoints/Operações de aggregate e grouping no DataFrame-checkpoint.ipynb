```python
# agrupar informações
df.groupby('candidate')
```

```python
# depois que fiz o agrupamento eu posso fazer agregações
# fazendo estatistica em cima de cada coluna
# neste caso irei passar um dicionario para a função aggregate 

# Sintaxe da função:
# df.groupby('candidate') = Agrupa os dados usando a coluna candidate como referencia
# .aggregate({'votes':[min, np.mean, max]}) = agregando os dados de votes ao candidate
df.groupby('candidate').aggregate({'votes':[min, np.mean, max]})
```

```python
# para saber as informações de um candidato especifico,
# por exemplo onde Hillary Clinton teve 590502 votos
df[df['votes'] == 590502]
```

```python
# agora eu quero agragar os dados de fraction_votes com candidates
df.groupby('candidate').aggregate({'fraction_votes':[min, np.mean, max]})
```

```python
# me traga os dados onde tenha as frações de votos (fraction_votes) = a 1
df[df['fraction_votes'] == 1]
```

```python
# Agora quero saber os dados que tiveram a fração de voto = a 1 e que sejam da Hillary Clinton
df[(df['fraction_votes'] == 1 ) & (df['candidate'] == 'Hillary Clinton')]
```

```python
# criando uma função que filtre pela coluna fraction_votes
def fraction_votes_filter(x):
    return x['votes'].sum() > 300000
df.groupby('state').filter(fraction_votes_filter)
```

```python
df[df['state_abbreviation'] == 'AL']['votes'].sum()
```

```python
# mais um exemplo
df.groupby(['state_abbreviation','candidate'])['votes'].sum()
```
