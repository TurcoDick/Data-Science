```python
import matplotlib.pyplot as plt
# a linha abaixo é necessario, mas não sei o porque ainda
%matplotlib inline
import numpy as np
```

```python
salario = np.array([100,200,300,500,400,150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='black')
```

```python
plt.plot(salario, c='black', ls='--')
```

```python
plt.plot(salario, c='black', ls='--', marker='s')
```

```python
# o plt.show tira aquela linha [<matplotlib.lines.Line2D at 0x7f7003f45e80>]
plt.plot(salario, c='black', ls='--',marker='s',ms=18)
plt.show()
```

```python
salarios_gileno = np.array([50,300,500,450,412,345])
plt.plot(salario, c='black', ls='--',marker='s',ms=5, label="Salários Marcos")
plt.plot(salarios_gileno, c='red', ls='--',marker='s',ms=5, label="Gileno")
plt.show()
```

```python
# colocando legenda
plt.plot(salario, c='blue', ls='--', marker='s', ms=5, label="Salarios Marcos")
plt.plot(salarios_gileno, c='red', ls='--', marker='s', ms = 5, label="Salários Gileno")
plt.legend()
plt.show()
```

```python
plt.plot(salario, c='blue', ls='--', marker='s', ms=5, label="Salarios Marcos")
plt.plot(salarios_gileno, c='red', ls='--', marker='^', ms = 5, label="Salários Gileno")
plt.legend()
plt.show()
```

```python
# mudando o lugar da legenda
plt.plot(salario, c='blue', ls='--', marker='s', ms=5, label="Salarios Marcos")
plt.plot(salarios_gileno, c='red', ls='--', marker='^', ms = 5, label="Salários Gileno")
plt.legend(loc ='upper left')
plt.show()
```
