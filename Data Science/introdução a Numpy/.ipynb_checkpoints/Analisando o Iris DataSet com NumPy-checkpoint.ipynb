```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plotando as 50 primeiras amostras
plt.plot(iris[:50,0],c='red', ls=':', marker='s', ms=5)
plt.show()
```

```python
# plotando os dados de 51 a 100
plt.plot(iris[50:100,0], c='black', ls='--', marker='s', ms=5 )
plt.show()
```

```python
plt.plot(iris[100:150,0], c='blue', ls='--', marker='o', ms=5)
plt.show()
```

```python
plt.plot(iris[:50,0],c='red', ls=':', marker='s', ms=5, label='Comp. Sepala iris-Setosa')
plt.plot(iris[50:100,0], c='black', ls='--', marker='s', ms=5,label='Comp. Sepala iris-Versicolor' )
plt.plot(iris[100:150,0], c='blue', ls='--', marker='o', ms=5, label='Comp. Sepala iris-Virginica')
plt.legend()
plt.show()
```
