```python
# a linha do with abre o arquivo original
# é importante que que eu mude o nome do arquivo de parametro no arq = open(''), se não o arquivo original será 
# alterado
with open('/home/alison/Projetos/Projeto-udemy-MachineLearning/DataSets/Datasets de treino/GSM2070024_sample_table.txt', 'r') as f:
    # a linha abaixo cria um novo arquivo com extensão .csv já com as mudanças que desejo
    arq = open('/home/alison/Projetos/Projeto-udemy-MachineLearning/DataSets/Datasets de treino/GSM2070024_sample_table.csv','a')
    for linha in f.readlines():
        arq.write(linha.replace('\t', ';'))
    arq.close()
```
