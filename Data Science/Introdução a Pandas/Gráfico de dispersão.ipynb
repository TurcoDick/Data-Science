```python
# outro exemplo
# agora com o dataset da iris
pydataset.data('iris', show_doc=True)
```

```python
iris = pydataset.data('iris')
```

```python
# vendo o inicio da tabela
iris.head()
```

```python
# vendo o fim da tabela 
iris.tail()
```

```python
# vamos usar 3 parametros 
# a parametro sizes significa = de aos pontos o tamanho proporcional aos valores desta coluna

plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], sizes = iris['Petal.Length'])
```

```python
# para almentar o tamanho dos pontos fazemos assim
# Interessante quando coloco plt.show(0) mostra dois graficos
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], sizes = 15 * iris['Petal.Length'])
plt.show(0)
```

```python
# criando uma coluna para usar cores
def specie_color(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    return 2
```

```python
iris['SpeciesNumber'] = iris['Species'].apply(specie_color)
```

```python
iris.head()
```

```python
iris.head(500)
```

```python
# oque esta configurando a cor dos pontos são as funções cmap='viridis' , alpha = 0.4
# preciso estudar isso

plt.figure(figsize=(15,10))
plt.scatter( iris['Sepal.Length'], iris['Sepal.Width'], sizes = 70 * iris['Petal.Length'], c= iris['SpeciesNumber'],
           cmap='viridis' , alpha = 0.4)

plt.show()
```

```python
# usando seaborn 
import seaborn as sns
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,10))
plt.scatter( iris['Sepal.Length'], iris['Sepal.Width'], sizes = 70 * iris['Petal.Length'], c= iris['SpeciesNumber'],
           cmap='viridis' , alpha = 0.4)

plt.show()
```

```python
# usando outros recursos do seaborn
# é mais facil usar o seaborn e para modificar o tamanho do grafico é só mudar o size, faz mais sentido não ? kkk
g = sns.lmplot(x="Sepal.Length", y="Sepal.Width", hue="Spe", truncate=True, size=10, data=iris)
g.set_axis_labels("Sepal length (mm)", "Sepal width (mm)")

plt.show()
```
